# Unificación y transformación de datos para uso en BI

```python
import pandas as pd
```

```python
data_original = pd.read_excel('Data_final_para_merge.xlsx')
data_etiquetas = pd.read_excel('etiquetas_finales.xlsx')
```

```python
mapeo = {0: 'Negativo', 1: 'Positivo', 2: 'Neutral'}
```

```python
data_etiquetas.keys()
```

```python
etiquetas = data_etiquetas[0].map(mapeo)
```

```python
data_original['etiqueta_clasificacion'] = etiquetas
```

```python
data_original.keys()
```

```python
data_original
```

```python
facultades = {
    'SALUD':'FACULTAD DE CIENCIAS DE LA SALUD',
    'ADMINISTRACION':'FACULTAD DE CIENCIAS ADMINISTRATIVAS ECONÓMICAS Y CONTABLES',
    'EDUCACION':'FACULTAD DE EDUCACIÓN',
    'AGROPECUARIAS':'FACULTAD DE CIENCIAS AGROPECUARIAS',
    'TRASVERSALES':'TRASVERSALES',
    'INGENIERIA':'FACULTAD DE INGENIERÍA',
    'SOCIALES':'FACULTAD DE CIENCIAS SOCIALES, HUMANIDADES Y CIENCIAS POLITICAS',
    'EDUCACION FISICA':'FACULTAD DE CIENCIAS DEL DEPORTE Y EDUCACIÓN FÍSICA',
}
```

```python
data_original['FACULTAD'] = data_original['FACULTAD'].map(facultades)
```

```python
data_original = data_original.drop('Unnamed: 0', axis = 1)
```
